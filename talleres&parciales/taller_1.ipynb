```python
from sympy import symbols, cos, sin, pi, integrate, simplify

# Variables simbólicas
t, A, B, w = symbols('t A B w', real=True, positive=True)

# Señal
x = A*cos(w*t) + B*sin(w*t)

# Periodo fundamental
T = 2*pi/w

# Potencia promedio = (1/T) * ∫_0^T x(t)^2 dt
x2 = x**2
potencia = (1/T) * integrate(x2, (t, 0, T))
potencia_simplificada = simplify(potencia)

# Mostrar resultado
print("Potencia promedio:", potencia_simplificada)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
a = 2
k = 1
t0 = 1
t1 = 3

# Tiempo
t = np.linspace(0, 5, 1000)

# Señal x(t)
x = np.where((t >= t0) & (t < t1), a * np.exp(k * t), 0)

# Graficar
plt.plot(t, x, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Señal x(t) = a * exp(k*t) entre t0 y t1')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np

# Señal 1: x[n] = n
N = 1000
n = np.arange(0, N+1)
x1 = n

energia1 = np.sum(np.abs(x1)**2)
potencia1 = np.mean(np.abs(x1)**2)

print("Señal 1: x[n] = n")
print("Energía:", energia1)
print("Potencia:", potencia1)

# Señal 2: x[n] = A*cos[n]*u[n - n0]
A = 2
n0 = 3
x2 = A * np.cos(n) * (n >= n0)

energia2 = np.sum(np.abs(x2)**2)
potencia2 = np.mean(np.abs(x2)**2)

print("\nSeñal 2: x[n] = A*cos[n]*u[n - n0]")
print("Energía:", energia2)
print("Potencia:", potencia2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Señal original
A_original = 30               # Amplitud original
omega = 100                   # Frecuencia angular
t = np.linspace(0, 0.1, 1000) # Tiempo continuo para la gráfica (100 ms)
x_continua = A_original * np.cos(omega * t)

# Paso 2: Ajustar la amplitud a ±3.3 V
A_max = 3.3
escala = A_max / A_original
x_escalada = escala * x_continua  # Ahora es 3.3*cos(100t)

# Paso 3: Muestreo (Fs > 2*15.9 ≈ 32 Hz, usamos por ejemplo Fs = 80 Hz)
Fs = 80                         # Frecuencia de muestreo
Ts = 1 / Fs                     # Período de muestreo
n = np.arange(0, 0.1, Ts)       # Instantes de muestreo en 100 ms
x_muestras = A_max * np.cos(omega * n)  # Señal muestreada

# Paso 4: Graficar
plt.figure(figsize=(10, 5))
plt.plot(t, x_escalada, label="Señal escalada (continua)", color='blue')
plt.stem(n, x_muestras, linefmt='red', markerfmt='ro', basefmt='k-', label="Muestras (digitalizadas)")
plt.title("Digitalización de $x(t) = 30\\cos(100t)$")
plt.xlabel("Tiempo [s]")
plt.ylabel("Voltaje [V]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
